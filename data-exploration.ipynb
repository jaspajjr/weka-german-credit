```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
raw_df = pd.read_csv('./credit-g.csv')

print(raw_df.info())

raw_df.sample(5)
```

```python
sns.set_style("whitegrid")

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16, 8))

sns.barplot([x for x in range(len(raw_df['property_magnitude'].value_counts()))],
            raw_df['property_magnitude'].value_counts(), ax=ax)

ax.set_xlabel('Predicted Class Label');
ax.set_ylabel('Count');
ax.set_xticklabels(raw_df['property_magnitude'].value_counts().index);
```

```python
sns.pairplot(raw_df)
```

```python
sns.set_style("whitegrid")

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16, 8))

sns.barplot([x for x in range(len(raw_df['class'].value_counts()))],
            raw_df['class'].value_counts(), ax=ax)

ax.set_xlabel('Credit Class Label');
ax.set_ylabel('Count');
ax.set_xticklabels(raw_df['class'].value_counts().index);
```
